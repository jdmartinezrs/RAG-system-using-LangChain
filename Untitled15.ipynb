```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_community.document_loaders import TextLoader, PyPDFLoader
from langchain_community.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA
from langchain.text_splitter import CharacterTextSplitter
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
import pandas as pd
import os
from google.colab import files
```

```python
# Asegúrate de que los nombres coincidan con los archivos que subiste
df1 = pd.read_excel("F-134_CONTROL_DE_MAQUINARIA_Y_VEHICULOS_ABRIL_2025 (2).xlsx", header=5)
df2 = pd.read_excel("F-73 Cronograma de Mantenimiento Preventivo 2023.xlsm", header=5)

# Convertir a texto plano
text1 = df1.to_string(index=False)
text2 = df2.to_string(index=False)

# Guardar como archivos de texto
with open("data1.txt", "w", encoding="utf-8") as f:
    f.write(text1)
with open("data2.txt", "w", encoding="utf-8") as f:
    f.write(text2)
```

```python
# Cargar los documentos
loader1 = TextLoader("data1.txt", encoding="utf-8")
loader2 = TextLoader("data2.txt", encoding="utf-8")
documents1 = loader1.load()
documents2 = loader2.load()

# PDFS (ajusta nombres si usas otros)
pdf1 = PyPDFLoader("P1-PLAN-DE-MANTENIMIENTO-PREVENTIVO-VEHICULOS.pdf").load()
pdf2 = PyPDFLoader("DISEÑO DE UN PLAN DE MANTENIMIENTO PREVENTIVO PARA LOS BRAZOS.pdf").load()
pdf3 = PyPDFLoader("E_K_GIZ_Manual-de-Conduccion-Eficiente.pdf").load()
pdf4 = PyPDFLoader("Cuerpo del trabajo PLAN DE MANTENIMIENTO PREVENTIVO DE VEHICULOS Y  MAQUINARIA PESADA PARA LA EMPRESA DE TRANSPORTE Y SERVICIOS DE COLOMBIA TRASERCOL S.A.S, UBICADA EN SAN MARTIN-CESAR. INFORME FINAL TRASERCOL_re.pdf").load()

# Combinar todos los documentos
all_documents = documents1 + documents2 + pdf1 + pdf2 + pdf3 + pdf4
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=400,  # reducir aún más para estar seguros
    chunk_overlap=50
)
docs = text_splitter.split_documents(all_documents)
```

```python
# Crear embeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Crear base vectorial (puede persistirse si lo deseas)
db = Chroma.from_documents(docs, embeddings)
retriever = db.as_retriever()
```

```python
model_id = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForSeq2SeqLM.from_pretrained(model_id)

pipe = pipeline("text2text-generation", model=model, tokenizer=tokenizer, max_length=512)
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)

query = "cada cuanto tiempo debo hacer mantenimiento a mi vehiculo?"
response = qa_chain.invoke({"query": query})

# Mostrar respuesta
print("📌 Respuesta:\n", response["result"])
print("\n📚 Documentos usados:")
fuentes = set()
for doc in response["source_documents"]:
    fuente = doc.metadata.get("source", "Desconocido")
    if fuente not in fuentes:
        fuentes.add(fuente)
        print("📄", fuente)
```
